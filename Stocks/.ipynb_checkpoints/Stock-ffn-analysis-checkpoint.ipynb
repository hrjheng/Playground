```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries, window, title_text):
    
    #Determing rolling statistics
    
    rolmean = pd.Series.rolling(timeseries, window=window).mean()
    rolstd = pd.Series.rolling(timeseries, window=window).std()
    #Plot rolling statistics:
    fig = plt.figure(figsize=(20,5))
    ax1 = fig.add_subplot(111,  ylabel='Price in $')

    ax1.plot(timeseries.index, timeseries, lw=2.5, label='Original')
    ax1.plot(timeseries.index, rolmean, lw=2.5, label='Rolling Mean')
    ax1.plot(timeseries.index, rolstd, lw=2.5, color='black', label = 'Rolling Std')
    ax1.set_xlabel('Date',fontsize=31)
    ax1.set_ylabel('Stock price', fontsize=31)
    ax1.tick_params(axis='x', labelsize=28)
    ax1.tick_params(axis='y', labelsize=28)
    # ax1.legend(loc='best',fontsize='large')
    ax1.legend(loc='best')
    
    plt.title(title_text,fontsize = 35)
    plt.show()
    
    #Perform Dickey-Fuller test:
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
```

```python
window = 60 
company = '2317tw'
test_stationarity(prices[company], window,'Rolling Mean & Standard Deviation')
```

```python
moving_avg = pd.Series.rolling(prices[company], window=window).mean()
plt.plot(prices[company])
plt.plot(moving_avg)
```

```python
# Subtract the rolling mean from the original series
moving_avg_diff = prices[company] - moving_avg
moving_avg_diff.dropna(inplace=True)
# moving_avg_diff.head()
test_stationarity(moving_avg_diff, window, 'Rolling Mean & Standard Deviation')
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(prices[company],freq=window)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

fig = plt.figure(figsize=(18,13))
# fig.suptitle('Company: TSM',fontsize = 25)
ax1 = fig.add_subplot(411)
ax2 = fig.add_subplot(412)
ax3 = fig.add_subplot(413)
ax4 = fig.add_subplot(414)

ax1.plot(prices[company].index, prices[company], lw=2.5,label='Original')
# ax1.legend(loc='best',fontsize='large')
ax1.legend(loc='best')
ax1.set_xlabel('Date',fontsize=26)
ax1.set_ylabel('Stock price', fontsize=26)
ax1.tick_params(axis='x', labelsize=23)
ax1.tick_params(axis='y', labelsize=23)
ax1.set_title('Company: '+company,fontsize = 35)

ax2.plot(prices[company].index,trend, lw=2.5,label='Trend',color='C1')
# ax2.legend(loc='best',fontsize='large')
ax2.legend(loc='best')
ax2.set_xlabel('Date',fontsize=26)
ax2.set_ylabel('Stock price', fontsize=26)
ax2.tick_params(axis='x', labelsize=23)
ax2.tick_params(axis='y', labelsize=23)

ax3.plot(prices[company].index,seasonal, lw=2.5,label='Seasonality',color='C2')
# ax3.legend(loc='best',fontsize='large')
ax3.legend(loc='best')
ax3.set_xlabel('Date',fontsize=26)
ax3.set_ylabel('Stock price', fontsize=26)
ax3.tick_params(axis='x', labelsize=23)
ax3.tick_params(axis='y', labelsize=23)

ax4.plot(prices[company].index,residual, lw=2.5,label='Residuals',color='C3')
# ax4.legend(loc='best',fontsize='large')
ax4.legend(loc='best')
ax4.set_xlabel('Date',fontsize=26)
ax4.set_ylabel('Stock price', fontsize=26)
ax4.tick_params(axis='x', labelsize=23)
ax4.tick_params(axis='y', labelsize=23)

plt.tight_layout()
plt.savefig('./plots/Stock-price-decomposition-'+company+'.png', bbox_inches = 'tight', dpi=500)
plt.savefig('./plots/Stock-price-decomposition-'+company+'.pdf', bbox_inches = 'tight')
```

```python
test_stationarity(seasonal, window, 'Seasonality')
```

```python
residual.dropna()
residual = residual[residual.notnull()]
test_stationarity(residual, window, 'Residual')
```
