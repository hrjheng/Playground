```python
import pandas_datareader.data as web
import matplotlib.pyplot as plt
from datetime import datetime
import pickle
```

## Fetch volatility ##

```python
with open('Monthly-revenue-201211to201911.pickle', 'rb') as file:
    data = pickle.load(file)
    
file.close()
```

```python
### 股價波動率 ###
def volatility(symbol):
    now = datetime.now()
    year = now.year
    month = now.month
    df = web.DataReader(symbol+'.TW', 'yahoo', datetime(year-7,month,1))
    dailyRet = df['Close'].pct_change()
    # dailyRet.plot(subplots = True, figsize = (10, 8))
    # plt.legend(loc='best')
    return dailyRet.std()

list_volatility = []
for sym in data['2019-11-01']['公司代號']:
    if sym == '全部國內上市公司合計':
        continue
        
    print ('company symbol: ',sym)
    list_volatility.append(volatility(sym))

# save the list of volatility to a pickle file
file = open('Volatility-7y.pickle', 'wb')
pickle.dump(list_volatility, file)
file.close()
```
